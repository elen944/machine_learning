```python
#Определить типы переменных

#номинальные переменные (не имеют порядка и сравнения)
#порядковые пиеременные (имеют порядок, но не имеют смысла мат операции)
#числовые непрерывные переменные (имеют порядок, есть смысл мат операций)
#Вывести статистику для числовых и для категоиральных перменных раздельно.
```

```python
ordinal_vars = ['Working Hours']  # можно рассматривать как порядковую

numeric_vars = ['id', 'age', 'Package', 'Salary Per Month']

# Предобработка числовых переменных (удаление единиц измерения)
df['Package'] = df['Package'].str.replace(' LPA', '').astype(float)
df['Salary Per Month'] = df['Salary Per Month'].str.replace(' Rs', '').astype(float)
df['Working Hours'] = df['Working Hours'].str.replace(' Hr', '').astype(int)

# Статистика для числовых переменных
print("=" * 60)
print("СТАТИСТИКА ДЛЯ ЧИСЛОВЫХ ПЕРЕМЕННЫХ")
print("=" * 60)

numeric_stats = df[numeric_vars + ['Working Hours']].describe()
print(numeric_stats)
```

```python
# Статистика для категориальных переменных
print("\n" + "=" * 60)
print("СТАТИСТИКА ДЛЯ КАТЕГОРИАЛЬНЫХ ПЕРЕМЕННЫХ")
print("=" * 60)

for col in nominal_vars:
    print(f"\n--- {col} ---")
    value_counts = df[col].value_counts()
    print(f"Количество уникальных значений: {df[col].nunique()}")
    print(f"Наиболее частые значения:")
    print(value_counts.head(10))  # Показываем топ-10 самых частых
```

```python
# Корреляционная матрица для числовых переменных
print("\n" + "=" * 50)
print("КОРРЕЛЯЦИОННАЯ МАТРИЦА (числовые переменные)")
print("=" * 50)

correlation_matrix = df[numeric_vars + ['Working Hours']].corr()
print(correlation_matrix.round(3))
```

```python
# Визуализация распределения числовых переменных
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for i, col in enumerate(numeric_vars + ['Working Hours']):
    df[col].hist(bins=30, ax=axes[i])
    axes[i].set_title(f'Распределение {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Частота')

plt.tight_layout()
plt.show()

# Визуализация топ-10 категорий для номинальных переменных
for col in nominal_vars[:6]:  # Показываем первые 6 для примера
    plt.figure(figsize=(10, 6))
    top_10 = df[col].value_counts().head(10)
    top_10.plot(kind='bar')
    plt.title(f'Топ-10 значений: {col}')
    plt.xlabel(col)
    plt.ylabel('Количество')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['age'])
plt.title('Боксплот возраста с выявлением выбросов')
plt.xlabel('Возраст (лет)')
plt.show()

# Статистика выбросов
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['age'] < lower_bound) | (df['age'] > upper_bound)]

print(f"Выбросы возраста: < {lower_bound:.1f} или > {upper_bound:.1f} лет")
print(f"Количество выбросов: {len(outliers)}")
print(f"Процент выбросов: {len(outliers)/len(df)*100:.1f}%")
```
